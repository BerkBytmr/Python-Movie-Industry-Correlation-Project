```python
#Importing Libraries

import pandas as pd
import seaborn as sns
import numpy as np 

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read in Data

df =pd.read_csv(r'C:\Users\HUAWEI\Downloads\movies.csv')



```

```python
# checking data

df.head()
```

```python
#checking if there is missing data

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))

```

```python
# data types for our columns

df.dtypes
df.head()

```

```python
# changing data type of columns 
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')

```

```python
# creating correct year column

df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')
```

```python
pd.set_option('display.max_rows',None)
```

```python
df=df.sort_values(by='gross',ascending=False,inplace=False)
df.head()
```

```python
# dropping any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates()
```

```python
# budget high corrrelation

# scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross',data=df, scatter_kws={"color": "red"},line_kws={"color": "blue"} )
plt.show()
```

```python
# correlation checking
df.corr(numeric_only=True,method='pearson') # because we have string columns in dataset
```

```python
# correlation matrix for numeric features
correlation_matrix = df.corr(numeric_only=True,method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
# changing string columns' type
df_copy = df.copy()
df_numberized = df_copy
for col_name in df_numberized:
    if(df_numberized[col_name].dtype=='object'):
        df_numberized[col_name]=df_numberized[col_name].astype('category')
        df_numberized[col_name] = df_numberized[col_name].cat.codes

df_numberized.head()

```

```python
# correlation matrix for all values ( because we turned string values to number format)
correlation_mat = df_numberized.corr(method='pearson')
sns.heatmap(correlation_mat,annot=True)
plt.show()
```

```python
correlation_mat=df_numberized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)
sorted_pairs
```

```python
# finding high correlations
high_corr = sorted_pairs[(sorted_pairs)>0.5].sort_values(ascending=False)
high_corr
```
